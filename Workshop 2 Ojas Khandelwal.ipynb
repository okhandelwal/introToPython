## Exercise 5: The Fibonacci Sequence

The Fibonacci sequence is a sequence of integers defined by the following recursion relation. The $n$-th integer $a_n$ is defined in terms of previous integers of the sequence as

$$a_n = a_{n-1} + a_{n-2}$$

and $a_0 = 0$ and $a_1 = 1$. So the first few numbers are 0, 1, 1, 2, 3, 5, 8, 13, 21, .... Write a program to print the Fibonacci series from 0 to 100.

```python
a0 = 0
a1 = 1
sum = 0
list = [a0,a1]
while True : 
    sum = a0 + a1
    a0 = a1
    a1 = sum
    if(sum>100):
        break
    list.append(sum)
print(list)
```

## Exercise 6: Prime Ribs

Write a function which returns a bool value True or False decided by whether or not the input value is a prime number.

```python

def isPrime(x):
    if (x%2 == 0):
        return False
    for i in range(3,x,2):
        if(x%i==0):
            return False
    return True


x = int(input("give a num: "))
bool = isPrime(x)
print(bool)
```

## Exercise 7: Escape Velocity

$$ V_{esc} = \left( \frac{2 G M}{R} \right)^{1/2} $$

6a) The formula for escape velocity is given above. In the cell below, create three different lists; one listing the names of each planet in the solar system, one listing their respective masses in kilograms, and one listing their radii in meters. Then, write the formula given above as a python function and use loops to calculate the escape velocity for each planet. 

```python
G = 6.6743e-11
names = ["Mercury", "Venus", "Earth", "Mars", "Jupiter", "Saturn", "Neptune", "Uranus"]
masses = [3.285e23, 4.867e24, 5.9722e24, 6.39e23, 1.898e27, 5.683e26, 1.024e26, 8.681e25]
radii = [2440, 6051, 6371, 3390, 69911, 58232, 25362, 24622]

def get_escape_velocity(mass, radius):
    return (2 * G * mass / radius)**0.5

for i in range(8):
    v = get_escape_velocity(masses[i], radii[i])
    print("Planet:", names[i], "   Mass: ", masses[i], "   Radius: ", radii[i], "     Escape velocity: {0:6.2f} m/s".format(v))
```

Then, in the second scatterplot, use the calculations you made in the previous part to plot escape velocity against both mass and radius in the same plot. Be sure to include a way for the reader of your plot to differentiate the two. Be sure to properly label all of your plots.

```python
import matplotlib.pyplot as plt

escapeV = []
for i in range(0,8):
     escapeV.append(get_escape_velocity(masses[i],radii[i]))
# Scatter Plot 1: Mass vs. Radius
plt.figure(figsize=(10, 6))
plt.scatter(masses, radii, color='blue', label='Mass vs Radius')
plt.xlabel('Mass (kg)')
plt.ylabel('Radius (m)')
plt.title('Mass vs. Radius of Planets in the Solar System')
plt.legend()
plt.grid(True)
plt.show()

# Scatter Plot 2: Escape Velocity vs. Mass and Radius
fig, ax1 = plt.subplots(figsize=(6, 8))
color = 'tab:blue'
ax1.set_ylabel('Mass (kg)',color=color)
ax1.set_xlabel('Escape Velocity (m/s)', )
ax1.scatter(escapeV, masses, color=color, label='Escape Velocity vs Mass')
ax1.tick_params(axis='y', labelcolor=color)
ax1.legend(loc='center left')

ax2 = ax1.twinx()
color = 'tab:red'
ax2.set_ylabel('Radius (m)', color=color)
ax2.scatter(escapeV, radii, color=color, label='Escape Velocity vs Radius')
ax2.tick_params(axis='y', labelcolor=color)
ax2.legend(loc='upper left')

plt.title('Escape Velocity vs. Mass and Radius of Planets in the Solar System')
plt.grid(True)
plt.show()
```

## Exercise 8: Particle in a Box

The energy levels for a quantum particle in a three-dimensional rectangular box of dimensions $\{L_1, L_2, \text{ and } L_3\}$ are given by

$$ E_{n_1, n_2, n_3} = \frac{\hbar^2 \pi^2}{2m} \left[ \frac{n_1^2}{L_1^2} + \frac{n_2^2}{L_2^2} + \frac{n_3^2}{L_3^2} \right] $$

where the $n$'s are integers greater than or equal to one. Your goal is to write a program that will calculate, and list in order of increasing energy, the values of the $n$'s for the 10 lowest *different* energy levels, given a box for which $L_2 = 2L_1$ and $L_3 = 4L_1$.

Your program should include two user-defined functions:

1. A function `energy(n1, n2, n3)` that takes integer values $n_1$, $n_2$, and $n_3$, and computes the corresponding energy level in units of $\hbar^2 \pi^2/2 m L_1^2$.

1. A function `lowest_unique_K(K, List)` which takes a positive integer $K$ and a list of real numbers `List`, and returns an ordered (ascending) list of the lowest $K$ unique numbers in the list `List`. For instance, `lowest_unique_K(3, [-0.5, 3, 3, 2, 6, 7, 7])` would return `[-0.5, 2, 3]`. The function should not modify the original list `List`.
    
You may also wish to make other user-defined functions depending on how you go about solving the problem. In fact, if you find some clever way to solve the problem that doesn't use `lowest_unique_K`, that is fine too! (You still need to write `lowest_unique_K`, though.) But whatever you do, be sure to comment your code clearly!

#### Output

Your final output should look like this (though with different numbers, and not necessarily the same number of lines):

    energy, n1, n2, n3
    (0.4375, 1, 1, 1)
    (0.625, 1, 2, 1)
    (0.8125, 2, 1, 1)
    (0.9375,  , 3, 1)
    (1.0, 2, 2, 1)
    (1.1875, 1, 1, 2)
    (1.3125, 2, 3, 1)
    (1.375, 1, 2, 2)
    (1.375, 1, 4, 1)
    (1.4375, 3, 1, 1)
    (1.5625, 2, 1, 2)

Notice how there are only 10 unique energies listed, but more than 10 lines.

```python
#energy
def energy(n1, n2, n3):
    return n1**2+n2**2/4+n3**2/16

#lowest_unique_K
def lowest_unique_K(K, List):
    new = []
    for num in List:
        if not value_in_list(num, new):
            new.append(num)
    return sorted(new)[:K]
#helper function for lowest_unique_K. Finds if the value of an object is in the list
def value_in_list(value, List):
    for item in List:
        if value == item:
            return True
    return False


# Generate a list of energy levels and their corresponding (n1, n2, n3) values
energy_levels = []
for n1 in range(1, 10):
    for n2 in range(1, 10):
        for n3 in range(1, 10):
            E = energy(n1, n2, n3)
            energy_levels.append((E, n1, n2, n3))

# Sort the energy levels by energy
energy_levels.sort()

# Extract the unique energy levels
unique_energies = []
unique_energy_set = set()
for E, n1, n2, n3 in energy_levels:
    if E not in unique_energy_set:
        unique_energies.append((E, n1, n2, n3))
        unique_energy_set.add(E)
    if len(unique_energies) == 10:
        break

# Print the results
print("energy, n1, n2, n3")
for E, n1, n2, n3 in unique_energies:
    print(f"({E}, {n1}, {n2}, {n3})")
```
