# Problem 3: Binomial Coefficients

The binomial coefficient $n \choose k$ is an integer equal to

$$ {n \choose k} = \frac{n!}{k!(n-k)!} = \frac{n \times (n-1) \times (n-2) \times \cdots \times (n-k + 1)}{1 \times 2 \times \cdots \times k} $$

when $k \geq 1$, or ${n \choose 0} = 1$ when $k=0$. (The special case $k=0$ can be included in the general definition by using the conventional definition $0! \equiv 1$.)

    - Write a function `heads_exactly(n,k)` to calculate the probability that a coin tossed $n$ times comes up heads exactly $k$ times.

Remember to write informative doc strings, comment your code, and use descriptive function and variable names so others (and future you) can understand what you're doing!


```python
import matplotlib.pyplot as plt

def factorial(n):  # Assuming n is greater than or equal to 0
    if n == 0:
        return 1
    result = 1
    for i in range(2, n + 1):
        result *= i
    return result

def binomial(n, k):
    if k < 0 or k > n:
        return 0  # Handle invalid values of k
    return factorial(n) // (factorial(k) * factorial(n - k))

def heads_exactly(n, k):
    return binomial(n, k) / 2**n

def heads_atleast(n, k):
    total_probability = 0
    for i in range(k, n + 1):
        total_probability += heads_exactly(n, i)
    return total_probability

# Example usage
n = 100
k = 60

exactly_probability = heads_exactly(n, k)
atleast_probability = heads_atleast(n, k)

print(f"The probability that a coin tossed {n} times comes up heads exactly {k} times is: {exactly_probability:.3f}")
print(f"The probability that a coin tossed {n} times comes up heads at least {k} times is: {atleast_probability:.3f}")

# Generate plots
k_values = list(range(n + 1))
exactly_probabilities = [heads_exactly(n, k) for k in k_values]
atleast_probabilities = [heads_atleast(n, k) for k in k_values]

# Plot heads_exactly
plt.figure(figsize=(14, 7))
plt.plot(k_values, exactly_probabilities, marker='o', linestyle='-', label='heads_exactly(n, k)')
plt.title('Probability of Exactly k Heads in 100 Tosses')
plt.xlabel('k')
plt.ylabel('Probability')
plt.legend()

# Plot heads_atleast
plt.figure(figsize=(14, 7))
plt.plot(k_values, atleast_probabilities,color = "r", marker='o', linestyle='-', label='heads_atleast(n, k)')
plt.title('Probability of At Least k Heads in 100 Tosses')
plt.xlabel('k')
plt.ylabel('Probability')
plt.legend()
```
