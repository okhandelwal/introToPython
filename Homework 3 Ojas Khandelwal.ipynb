1a) Write a program that produces a random number and prints heads for even numbers and tails for odd numbers.

1b) Write a Python program which takes two digits m (row) and n (column) as input and generates a two-dimensional array. The element value in the i-th row and j-th column of the array should be i*j.

Note :
i = 0,1.., m-1
j = 0,1, n-1.

```python
m = input()
n = input()
output = []
for i in range (int(m)):
    output.append([])
    for j in range(int(n)):
        output[len(output)-1].append(i*j)
print(output)
```

1c) The rate at which temperature changes with height is dependent on where you are in the atmosphere. Look up the temperature gradient functions for the different levels of the atmosphere in a standard day and write a python function which takes height as an input and outputs temperature.

```python

#not sure if I got the right equations as part of the data, I sorta just searched as much as I could/utilized chat gpt to give me info
#on temperature through different layers but im not positive I got it right.

def temperature_at_height(height):
    height = int(height)
    if height < 0:
        return  
    if height <= 11:
        # Troposphere
        T = 288.15 - 6.5 * height
    elif height <= 47:
        # Stratosphere
        if height <= 20:
            T = 216.65
        elif height <= 32:
            T = 216.65 + 1 * (height - 20)
        else:
            T = 228.65 + 2.8 * (height - 32)
    elif height <= 86:
        # Mesosphere
        if height <= 51:
            T = 270.65
        elif height <= 71:
            T = 270.65 - 2.8 * (height - 51)
        else:
            T = 214.65 - 2 * (height - 71)
    else:
        # Thermosphere (simplified, linear increase)
        T = 214.65 - 2 * (86 - 71) + 3.0 * (height - 86)
    
    return T

height = input()
temperature = temperature_at_height(height)
print(f"The temperature at {height} km is {temperature} K")
```

Write a python function for sinc(x). Make sure that your function handles $x=0$ correctly. Plot your result.

2c) The outer and inner limits of the habitable zone for a stellar system can be described using Luminosity in the equations below. Finally, add these values to your dataframe, adding to the useful data you have easy access to.

```python
def calcinner(L):
    return np.sqrt(L/1.1) #in AU
def calcupper(L):
    return np.sqrt(L/.53) #in AU
pl_lower = df['st_lum'].apply(calcinner)
pl_upper = df['st_lum'].apply(calcupper)
df['pl_lower'] = pl_lower
df['pl_upper'] = pl_upper
df_refined = df.dropna(subset=['pl_lower','pl_upper'])
df_refined.plot.scatter('st_lum','pl_upper')
df_refined.plot.scatter('st_lum','pl_lower',c = "r")
```
