## Monte Carlo Techniques

### Example: coin flip

```python
outcomes = ('Heads','Tails')

def flip():
    if np.random.random() > 0.5:
        return outcomes[1]
    
    return outcomes[0]

heads = 0
tails = 0
count = 0
while count < 100000:
    outcome = flip()
    if outcome == 'Heads':
        heads +=1
    else:
        tails +=1
    count += 1

print(heads, tails)
```

### Example: dice roll

```python
def dice():
    return int(np.random.random()*6)+1

for i in range(10):
    print (dice())


x = [dice() for i in range(100000)]
plt.hist(x,6,(0.5,6.5));
plt.xlabel('Outcome of Roll')
plt.ylabel('Frequency of the outcome')

print(np.mean(x))
```

### Example: Deck of Cards

```python
import random
import itertools
SUITS = 'cdhs'
RANKS = '23456789TJQKA'
DECK = tuple(''.join(card) for card in itertools.product(RANKS, SUITS))
hand = random.sample(DECK, 10)
print (hand)
```

### Example:  Linear Congruent Generator

```python
myRandomSeed = 343
def myRandom(a=65539, b=0, c=int(2**31-1)):
    global myRandomSeed
    x = (a*myRandomSeed+b) % c
    myRandomSeed = x
    return x

out = ""
for i in range(5):
    out += str(myRandom())
    out += " "
    
print (out)
    
out = ""
myRandomSeed = 1
for i in range(20):
    out += str(myRandom(a=5, b=3, c=8))
    out += " "

print (out)
```

```python
# plot distribution
ds = np.random.normal(5,10,size=10000000)
print('Mean = {0:5.3f}'.format(np.mean(ds)))
print('Std. dev. = {0:5.3f}'.format(np.std(ds)))
plt.hist(ds,50)
plt.show()
```

### Example: Python random number generators

```python
ds = np.random.exponential(scale=2,size=10000)
print('Mean = {0:5.3f}'.format(np.mean(ds)))
print('Std. dev. = {0:5.3f}'.format(np.std(ds)))
plt.hist(ds,50)
plt.show()
```

```python
ds = np.random.standard_cauchy(size=100)
print('Mean = {0:5.3f}'.format(sp.mean(ds)))
print('Std. dev. = {0:5.3f}'.format(sp.std(ds)))
plt.hist(ds,20)
plt.show()
```

```python
ds = np.random.triangular(5,10,15,size=1000000)
print('Mean = {0:5.3f}'.format(sp.mean(ds)))
print('Std. dev. = {0:5.3f}'.format(sp.std(ds)))
plt.hist(ds,50)
plt.show()
```
